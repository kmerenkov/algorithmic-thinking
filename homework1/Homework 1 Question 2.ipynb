```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 10, 6
```

```python
import random


def generate_random_graph(degree, probability):
    graph = {}
    for i in range(0, degree):
        for j in range(0, degree):
            if i == j:
                continue
            if i not in graph:
                graph[i] = set()
            roll = random.random()
            if roll < probability:
                graph[i].add(j)
    return graph
```

```python
def compute_in_degrees(digraph):
    """Takes a directed graph digraph (represented as a dictionary)
    and computes the in-degrees for the nodes in the graph"""
    graph = {}
    for node1 in digraph:
        graph[node1] = 0
    for nodes in digraph.itervalues():
        for node in nodes:
            graph[node] += 1
    return graph


def in_degree_norm_distribution(digraph):
    """Computes normalized indegree distribution of a digraph.
    """
    indeg = compute_in_degrees(digraph)
    result = {}
    sum_indegrees = 0
    for val in indeg.itervalues():
        if val not in result:
            result[val] = 0
        result[val] += 1
        sum_indegrees += 1
    for val in result.iterkeys():
        result[val] = result[val] / float(sum_indegrees)
    return result
```

```python
random_graph_1 = generate_random_graph(3000, 0.9)
random_graph_2 = generate_random_graph(3000, 0.6)
random_graph_3 = generate_random_graph(3000, 0.2)
```

```python
plt.xscale("log")
plt.yscale("log")

import matplotlib.patches as mpatches


plots = [
         ("green", "g.", "3000 nodes with p=0.9", random_graph_1),
         ("blue", "b.", "3000 nodes with p=0.6", random_graph_2),
         ("red", "r.", "3000 nodes with p=0.2", random_graph_3),
         ("cyan", "c.", "citation papers graph", citation_graph),
]

handles = [mpatches.Patch(color=color, label=label) for color, _1, label, _4 in plots]

for color, options, label, graph in plots:
    distribution = in_degree_norm_distribution(graph)
    
    xs, ys = zip(*distribution.items())
    plt.plot(xs, ys, options, label=label)

plt.legend()

plt.show()
```
