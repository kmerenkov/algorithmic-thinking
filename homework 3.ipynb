```python
def mystery(a, l, r):
    """sorted array of distinct integers a, left/right boundaries l/r"""
    print "="*80
    print "l =", l, "r =", r
    print "subset", a[l:r]
    if l > r:
        return -1
    m = int((l+r)/2.0)
    print "m =", m
    print "a[m] =", a[m]
    if a[m] == m:
        return m
    elif a[m] < m:
        return mystery(a, m+1, r)
    else:
        return mystery(a, l, m-1)
    
mystery(range(1, 10), 0, 9)
```
