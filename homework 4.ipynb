```python
from itertools import product
from collections import namedtuple
from pprint import pprint


ScoringMatrix = namedtuple('ScoringMatrix', ['match', 'mismatch', 'left_match', 'right_match'])
SCORING_MATRIX = ScoringMatrix(5, 2, -2, -4)


def global_alignment(string1, string2, scoring_matrix):
    alignment_matrix = compute_global_alignment_scores(string1, string2, scoring_matrix)
    align1, align2 = compute_global_alignment(string1, string2, scoring_matrix, alignment_matrix)
    return align1, align2


def local_alignment(string1, string2, scoring_matrix):
    alignment_matrix = compute_global_alignment_scores(string1, string2, scoring_matrix, local=True)
    align1, align2 = compute_global_alignment(string1, string2, scoring_matrix, alignment_matrix)
    return align1, align2


def compute_global_alignment_scores_222(string1, string2, scoring_matrix):
    def score_substrings(substr1, substr2):
        substr1 = substr1.rjust(len(substr2), '-')
        substr2 = substr2.rjust(len(substr1), '-')
        return sum(map(lambda x: score_chars(x[0], x[1]), zip(substr1, substr2)))
    def score_chars(c1, c2):
        if c1 == c2:
            return scoring_matrix.match
        if c1 and c2 == '-':
            return scoring_matrix.left_match
        if c2 and c1 == '-':
            return scoring_matrix.right_match
        if c1 != c2:
            return scoring_matrix.mismatch
        1/0  # XXX

    alignment_matrix = []
    for i in range(len(string1)):
        alignment_matrix.append([])
        for j in range(len(string2)):
            alignment_matrix[i].append(0)
    for i, j in product(range(1, len(string1)), range(1, len(string2))):
        c1 = string1[i]
        c2 = string2[j]
        score = score_chars(c1, c2)
        alignment_matrix[i-1][j-1] = score_substrings(string1[:i-1], string2[:j-1])
        alignment_matrix[i-1][j] = score_substrings(string1[:i-1], string2[:j])
        alignment_matrix[i][j-1] = score_substrings(string1[:i], string2[:j-1])
        alignment_matrix[i][j] = score_substrings(string1[:i], string2[:j])
    pprint(alignment_matrix)
    return alignment_matrix


def compute_global_alignment_scores(string1, string2, scoring_matrix, local=False):
    def score_chars(c1, c2):
        if c1 == c2:
            return scoring_matrix.match
        if c1 and c2 == '-':
            return scoring_matrix.left_match
        if c2 and c1 == '-':
            return scoring_matrix.right_match
        if c1 != c2:
            return scoring_matrix.mismatch
        1/0  # XXX

    m = len(string1)
    n = len(string2)
    alignment_matrix = []
    for i in range(m+1):
        alignment_matrix.append([])
        for j in range(n+1):
            alignment_matrix[i].append(0)
    alignment_matrix[0][0] = 0
    for i in range(1, m+1):
        alignment_matrix[i][0] = alignment_matrix[i-1][0] + scoring_matrix.left_match
    for j in range(1, n+1):
        alignment_matrix[0][j] = alignment_matrix[0][j-1] + scoring_matrix.right_match
    for i in range(1, m+1):
        for j in range(1, n+1):
            score = max([alignment_matrix[i-1][j-1] + score_chars(string1[i-1], string2[j-1]),
                         alignment_matrix[i-1][j] + scoring_matrix.left_match,
                         alignment_matrix[i][j-1] + scoring_matrix.right_match,])
            if not local:
                alignment_matrix[i][j] = score
            else:
                alignment_matrix[i][j] = 0 if score < 0 else score
    pprint(alignment_matrix)
    return alignment_matrix


def compute_global_alignment(string1, string2, scoring_matrix, alignment_matrix):
    def score_chars(c1, c2):
        if c1 == c2:
            return scoring_matrix.match
        if c1 and c2 == '-':
            return scoring_matrix.left_match
        if c2 and c1 == '-':
            return scoring_matrix.right_match
        if c1 != c2:
            return scoring_matrix.mismatch
        1/0  # XXX


    i = len(string1)
    j = len(string2)
    alignment1 = []
    alignment2 = []
    while i and j:
        if alignment_matrix[i][j] == alignment_matrix[i-1][j-1] + score_chars(string1[i-1], string2[j-1]):
            alignment1.insert(0, string1[i-1])
            alignment2.insert(0, string2[j-1])
            i = i - 1
            j = j - 1
        else:
            if alignment_matrix[i][j] == alignment_matrix[i-1][j] + scoring_matrix.left_match:
                alignment1.insert(0, string1[i-1])
                alignment2.insert(0, '-')
                i = i - 1
            else:
                alignment1.insert(0, '-')
                alignment2.insert(0, string2[j-1])
                j = j - 1
    while i:
        alignment1.insert(0, string1[i-1])
        alignment2.insert(0, '-')
        i = i - 1
    while j:
        alignment1.insert(0, '-')
        alignment2.insert(0, string2[j-1])
        j = j - 1
    return ''.join(alignment1), ''.join(alignment2)


#compute_global_alignment_scores('AC', 'TAG', SCORING_MATRIX)
global_alignment('AC', 'TAG', SCORING_MATRIX)
```

```python
# question 11
def find_substrings(string):
    for i in range(len(string)):
        for j in range(1, len(string)+1):
            if i >= j:
                continue
            yield string[i:j]
    yield ""


print list(find_substrings("AB"))
print list(find_substrings("ABC"))
print list(find_substrings("ABCD"))
```
