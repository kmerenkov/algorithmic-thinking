```python
import itertools
import pickle
import random

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 10, 6


def compute_degree(ugraph):
    """Takes a undirected graph graph (represented as a dictionary)
    and computes the degrees for the nodes in the graph"""
    graph = {node: 0 for node in ugraph}
    for nodes in ugraph.itervalues():
        for node in nodes:
            graph[node] += 1
    return graph


def generate_random_ugraph(degree, probability):
    graph = {}
    for i in range(0, degree):
        for j in range(0, degree):
            if i >= j:
                continue
            if i not in graph:
                graph[i] = set()
            if j not in graph:
                graph[j] = set()
            roll = random.random()
            if roll < probability:
                graph[i].add(j)
                graph[j].add(i)
    return graph


def generate_complete_ugraph(num_of_nodes):
    graph = {}
    for i in range(0, num_of_nodes):
        for j in range(0, num_of_nodes):
            if i == j:
                continue
            if i not in graph:
                graph[i] = set()
            if j not in graph:
                graph[j] = set()
            graph[i].add(j)
            graph[j].add(i)
    return graph


def upa(num_of_nodes, m):
    """Generate undirected graph"""
    assert 1 <= m <= num_of_nodes
    graph = generate_complete_ugraph(m)
    deg = compute_degree(graph)
    nodes_ps = list(itertools.chain.from_iterable([node] * (deg[node] + 1) for node in graph.keys()))
    for i in range(m, num_of_nodes):
        nodes = set()
        for _ in range(m):
            node = random.choice(nodes_ps)
            nodes.add(node)
        graph[i] = set()
        for node in nodes:
            if node not in graph:
                graph[node] = set()
        nodes_ps += list(nodes)
        nodes_ps.append(i)
        for node in nodes:
            graph[i].add(node)
            graph[node].add(i)
    return graph


def total_degree(graph):
    total_deg = sum(map(len, graph.itervalues()))
    return total_deg / 2


def possible_edges_count(n):
    return (n*(n-1)) / 2


def copy_graph(graph):
    """
    Make a copy of a graph
    """
    new_graph = {}
    for node in graph:
        new_graph[node] = set(graph[node])
    return new_graph


def delete_node(ugraph, node):
    """
    Delete a node from an undirected graph
    """
    neighbors = ugraph[node]
    ugraph.pop(node)
    for neighbor in neighbors:
        ugraph[neighbor].remove(node)


def bfs_visited(graph, node):
    """BFS visited"""
    queue = [node]
    visited = {node}
    while queue:
        node = queue.pop(0)
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    return visited


def copy_graph(graph):
    """
    Make a copy of a graph
    """
    new_graph = {}
    for node in graph:
        new_graph[node] = set(graph[node])
    return new_graph


def cc_visited(graph):
    """CC visited"""
    remaining_nodes = list(graph.keys())
    components = []
    while remaining_nodes:
        node = remaining_nodes[0]
        component = bfs_visited(graph, node)
        if component not in components:
            components.append(component)
        #remaining_nodes = list(set(remaining_nodes) - {node})
        remaining_nodes = list(set(remaining_nodes) - component)
    return components


def largest_cc_size(graph):
    """Largest CC size"""
    ccs = cc_visited(graph)
    if not ccs:
        return 0
    return max(len(x) for x in ccs)


def compute_resilience(graph, attack_order):
    """Resilience"""
    graph = copy_graph(graph)
    resilience = [largest_cc_size(graph)]
    for node in attack_order:
        del graph[node]
        for neighbors in graph.itervalues():
            if node in neighbors:
                neighbors.remove(node)
        resilience.append(largest_cc_size(graph))
    return resilience


def targeted_order(ugraph):
    """
    Compute a targeted attack order consisting
    of nodes of maximal degree
    
    Returns:
    A list of nodes
    """
    # copy the graph
    new_graph = copy_graph(ugraph)
    
    order = []    
    while len(new_graph) > 0:
        max_degree = -1
        for node in new_graph:
            if len(new_graph[node]) > max_degree:
                max_degree = len(new_graph[node])
                max_degree_node = node
        
        neighbors = new_graph[max_degree_node]
        new_graph.pop(max_degree_node)
        for neighbor in neighbors:
            new_graph[neighbor].remove(max_degree_node)

        order.append(max_degree_node)
    return order


def random_order(graph):
    nodes = list(graph)
    random.shuffle(nodes)
    return nodes


def degree_distribution(ugraph):
    """Takes a undirected graph ugraph (represented as a dictionary)
    and computes the unnormalized distribution of the degrees of the graph"""
    d_graph = compute_degree(ugraph)
    distribution = {}
    for val in d_graph.itervalues():
        if val not in distribution:
            distribution[val] = 0
        distribution[val] += 1
    return distribution


def group_by_degree(d_graph):
    graph = {}
    key = lambda x: x[1]
    for k, g in itertools.groupby(sorted(d_graph.items(), key=key), key=key):
        graph[k] = set(x[0] for x in g)
    return graph
```

```python
graph = {0: {1, 2},
         1: {0},
         2: {0},
         3: {},
}


def fast_targeted_order(graph):
    graph = copy_graph(graph)

    degree_sets = {}
    for node, neighbors in graph.iteritems():
        degree = len(neighbors)

        if degree not in degree_sets:
            degree_sets[degree] = {node}
        else:
            degree_sets[degree].add(node)

    L = []
    #for k in range(len(graph)-1, -1, -1):
    for k in reversed(range(len(graph))):
        while degree_sets.get(k):
            node = degree_sets[k].pop()
            if node not in graph:
                continue
            for neighbor in graph[node]:
                degree = len(graph[node])
                if degree in degree_sets:
                    degree_sets[degree] = degree_sets[degree] - {neighbor}
                if degree > 0:
                    if degree-1 not in degree_sets:
                        degree_sets[degree-1] = set()
                    degree_sets[degree-1].add(neighbor)
            L.append(node)
            delete_node(graph, node)
    return L


def degree(node):
    return len(network_graph[node])


FTO = fast_targeted_order(network_graph)
TO = targeted_order(network_graph)
from pprint import pprint
if len(FTO) != len(TO):
    print "="*80
    print len(FTO), "vs", len(TO)
    pprint(zip(map(degree, FTO)[:40], map(degree, TO)[:40]))
    print "="*80
```
