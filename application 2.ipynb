```python
import itertools
import pickle
import random

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 10, 6


def compute_in_degrees(digraph):
    """Takes a directed graph digraph (represented as a dictionary)
    and computes the in-degrees for the nodes in the graph"""
    graph = {}
    for node1 in digraph:
        graph[node1] = 0
    for nodes in digraph.itervalues():
        for node in nodes:
            graph[node] += 1
    return graph


def generate_random_ugraph(degree, probability):
    graph = {}
    for i in range(0, degree):
        for j in range(0, degree):
            if i >= j:
                continue
            if i not in graph:
                graph[i] = set()
            if j not in graph:
                graph[j] = set()
            roll = random.random()
            if roll < probability:
                graph[i].add(j)
                graph[j].add(i)
    return graph


def generate_complete_ugraph(num_of_nodes):
    graph = {}
    for i in range(0, num_of_nodes):
        for j in range(0, num_of_nodes):
            if i == j:
                continue
            if i not in graph:
                graph[i] = set()
            if j not in graph:
                graph[j] = set()
            graph[i].add(j)
            graph[j].add(i)
    return graph


def upa(num_of_nodes, m):
    """Generate undirected graph"""
    assert 1 <= m <= num_of_nodes
    graph = generate_complete_ugraph(m)
    indeg = compute_in_degrees(graph)
    nodes_ps = list(itertools.chain.from_iterable([node] * (indeg[node] + 1) for node in graph.keys()))
    for i in range(m, num_of_nodes):
        nodes = set()
        for _ in range(m):
            node = random.choice(nodes_ps)
            nodes.add(node)
        graph[i] = set()
        for node in nodes:
            if node not in graph:
                graph[node] = set()
        nodes_ps += list(nodes)
        nodes_ps.append(i)
        for node in nodes:
            graph[i].add(node)
            graph[node].add(i)
    return graph


def total_degree(graph):
    total_deg = sum(map(len, graph.itervalues()))
    return total_deg / 2


def possible_edges_count(n):
    return (n*(n-1)) / 2


def copy_graph(graph):
    """
    Make a copy of a graph
    """
    new_graph = {}
    for node in graph:
        new_graph[node] = set(graph[node])
    return new_graph

def delete_node(ugraph, node):
    """
    Delete a node from an undirected graph
    """
    neighbors = ugraph[node]
    ugraph.pop(node)
    for neighbor in neighbors:
        ugraph[neighbor].remove(node)


def random_order(graph):
    nodes = list(graph)
    random.shuffle(nodes)
    return nodes
```
